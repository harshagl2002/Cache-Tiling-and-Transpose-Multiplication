# 1.1(1) Benchmark





```python
import matplotlib.pyplot as plt

# Data
tile_size = [2, 4, 8, 16, 32, 64, 128, 256, 512]
time_in_ns = [23881191, 8935066, 8849118, 13001709, 13813448, 17077481, 20021126, 19831766, 19110795]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(tile_size, time_in_ns, marker='o', color='blue', linestyle='-')

# Add title and labels
plt.title('Tile Size vs Time (in ns)')
plt.xlabel('Tile Size')
plt.ylabel('Time (ns)')

# Add grid
plt.grid(True)

# Show the plot
plt.show()
```

## 1.3 OpenMP Parallelization

# Without using omp parallel for - Squares

```python
#without opm parallel for
num_threads = [1, 2, 4]
time = [76196, 80201, 80199]

# Create the plot
plt.figure(figsize=(5, 3))
plt.plot(num_threads, time, marker='o', color='blue', linestyle='-')

# Add title and labels
plt.title('Num threads vs Time')
plt.xlabel('Number of Threads')
plt.ylabel('Time (ns)')

#Add grid
plt.grid(True)

# Show the plot
plt.show()
```
